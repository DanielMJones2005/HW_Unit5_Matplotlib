## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame
treatment_data_mean = treatment_data_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean().reset_index()
t_data_mean_df = pd.DataFrame(treatment_data_mean)
t_data_mean_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame

treatment_data_sem = treatment_data_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem().reset_index()
t_data_sem_df = pd.DataFrame(treatment_data_sem)
t_data_sem_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
#t_data_mean_df
t_data_mean_dfpvt = t_data_mean_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")
t_data_mean_dfpvt
```

```python
sem_mean = t_data_sem_df.groupby(["Drug"])["Tumor Volume (mm3)"].mean()
```

```python
# Generate the Plot (with Error Bars)
x = t_data_mean_dfpvt.index

y_columns = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]

yerr_sem1 = (sem_mean.loc[["Capomulin"]])
yerr_sem2 = (sem_mean.loc[["Infubinol"]])
yerr_sem3 = (sem_mean.loc[["Ketapril"]])
yerr_sem4 = (sem_mean.loc[["Placebo"]])

plt.errorbar(x = x, y = t_data_mean_dfpvt["Capomulin"], yerr=int(yerr_sem1)^2, ecolor="red", color="red", linestyle=(0, (.5,.5)), marker="o")
plt.errorbar(x = x, y = t_data_mean_dfpvt["Infubinol"], yerr=int(yerr_sem2)^2, ecolor="blue", color="blue", linestyle=(0, (.5,.5)), marker="^")
plt.errorbar(x = x, y = t_data_mean_dfpvt["Ketapril"], yerr=int(yerr_sem3)^2, ecolor="green", color="green", linestyle=(0, (.5,.5)), marker="s")
plt.errorbar(x = x, y = t_data_mean_dfpvt["Placebo"], yerr=int(yerr_sem4)^2, ecolor="black", color="black", linestyle=(0, (.5,.5)), marker="d")
plt.legend(y_columns)

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

plt.xticks([0, 10, 20, 30, 40])
plt.xlim(-2,47)
plt.ylim(33,74)
plt.grid()

# Save the Figure
plt.savefig('1_TumorResponseTreatment.png') 
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
metas_data_mean = treatment_data_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean()
metas_data_mean_df = pd.DataFrame(metas_data_mean)
metas_data_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
metas_data_sem = treatment_data_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem()
metas_data_sem_df = pd.DataFrame(metas_data_sem)
metas_data_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
metas_data_mean2 = treatment_data_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean().reset_index()
metas_data_mean2_df = pd.DataFrame(metas_data_mean2)
metas_data_mean2_df
metas_data_mean_dfpvt = metas_data_mean2_df.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")
metas_data_mean_dfpvt
```

```python
metas_sem_mean = metas_data_sem_df.groupby(["Drug"])["Metastatic Sites"].mean()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
# Preview DataFrame
mice_count = treatment_data_df.groupby(["Drug","Timepoint"])["Mouse ID"].count().reset_index()
mice_count_df = pd.DataFrame(mice_count)
mice_count_df = mice_count_df.rename(columns={"Mouse ID": "Mouse Count"
            })
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview the Data Frame

mice_count_dfpvt = mice_count_df.pivot(index="Timepoint", columns="Drug", values="Mouse Count")
mice_count_dfpvt.head()
```

```python
# Generate the Plot (Accounting for percentages)
x = mice_count_dfpvt.index

y1 = (mice_count_dfpvt["Capomulin"]/25)*100
y2 = (mice_count_dfpvt["Infubinol"]/25)*100
y3 = (mice_count_dfpvt["Ketapril"]/25)*100
y4 = (mice_count_dfpvt["Placebo"]/25)*100

y_columns = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]

plt.plot(x, y1, color="red", linestyle=(0, (.5,.5)), marker="o")
plt.plot(x, y2, color="blue", linestyle=(0, (.5,.5)),marker="^")
plt.plot(x, y3, color="green", linestyle=(0, (.5,.5)),marker="s")
plt.plot(x, y4, color="black", linestyle=(0, (.5,.5)),marker="d")

plt.legend(y_columns)

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Surival Rate (%)")

plt.xticks([0, 10, 20, 30, 40])
plt.xlim(-2,47)
plt.ylim(33,103)
plt.grid()

# Save the Figure
plt.savefig('3_SurvivalDuringTreatment.png') 
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
drug_names = [t_data_mean_dfpvt.columns]
drug_names_df = pd.DataFrame(drug_names)
vol_per_chng_df = drug_names_df.transpose() 
vol_per_chng_df = vol_per_chng_df.rename(columns={0: "Drug"
            })

first_row = t_data_mean_dfpvt.iloc[[0]].values[0]
last_row = t_data_mean_dfpvt.iloc[[-1]].values[0]

tum_vol_per_chng = np.round(((last_row - first_row) / first_row)*100,6)

tum_vol_per_chng_df = pd.DataFrame(tum_vol_per_chng)
tum_vol_per_chng_df = tum_vol_per_chng_df.rename(columns={0: "% Change"
            })
tum_vol_per_chng_df
drug_tum_per_chng = pd.merge(vol_per_chng_df, tum_vol_per_chng_df, left_index=True, right_index=True)

# Display the data to confirm
drug_tum_per_chng
```

```python
# Store all Relevant Percent Changes into a Tuple


drugs_view = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
drugs_analyzed = drug_tum_per_chng.loc[drug_tum_per_chng["Drug"].isin(drugs_view)].reset_index()
drugs_analyzed_df = drugs_analyzed.drop(["index"], axis=1)

x_ax1 = drugs_analyzed_df["Drug"][0]
y_ax1 = drugs_analyzed_df["% Change"][0]

x_ax2 = drugs_analyzed_df["Drug"][1:]
y_ax2 = drugs_analyzed_df["% Change"][1:]

# Splice the data between passing and failing drugs
plt.bar(x_ax1, y_ax1, color="green", width=-1, align="edge") 
plt.bar(x_ax2, y_ax2, color="red", width=-1, align="edge") 

# Orient widths. Add labels, tick marks, etc. 
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.yticks([-20, 0, 20, 40, 60])
plt.ylim(-30,70)
plt.grid()

# Use functions to label the percentages of changes
def drug_perc_label(drug_x):
    drug_percent = int((drugs_analyzed_df[drugs_analyzed["Drug"]==drug_x]["% Change"].values))
    drug_label = (str(drug_percent) + "%")
        
    
    idx_pos = drugs_analyzed[drugs_analyzed_df["Drug"]==drug_x].index[0]
    #print(idx_pos)
    
    if idx_pos == 0:
        horiz = -.7
    elif idx_pos > 0:
        horiz = -.65
        horiz = horiz + idx_pos

    if drug_percent < 0:
        vert = -8
    elif drug_percent > 0:
        vert = 4
        
    plt.text(horiz, vert, drug_label, color="white", fontweight="bold")
    
# Call functions to implement the function calls
drug_perc_label("Capomulin")
drug_perc_label("Infubinol")
drug_perc_label("Ketapril")
drug_perc_label("Placebo")

# Save the Figure
plt.savefig('4_TumorChangeOver45DayTreatment.png') 

# Show the Figure
plt.show()
```
